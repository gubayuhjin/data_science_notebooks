I didn't find a gini function in Python library. So I wrote my own function :-)

```python
import numpy as np

def gini_coeff(x):
    n = len(x)
    s = x.sum()
    r = np.argsort(np.argsort(-x))
    return 1 - (2.0 * (r*x).sum() + s)/(n*s)

gini_coeff(data_metro.CENSUS2010POP)
```
